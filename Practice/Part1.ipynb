## 3-3. 예외 처리 구문 (Try / Except / Finally)

- 특정 구문을 수행하다가 구조적인 오류들이 발생
- 오류를 방지하기 위해 사용하는 예외 처리 구문
- 프로그램 형태로 파이썬 코드를 수행시킬 때 많이 사용
- 데이터 분석에서는 크롤링 또는 업무자동화(RPA)같은 자동화된 반복 수행 작업에 필수로 사용

```python
try :
    data14 = []
    for i in range(0,20):
        data14.append(i)
    data14.remove(20)
# try구문이 정상적으로 실행되다가 오류가 발생할 경우 아래 종속문장을 수행
except Exception as e: 
# Exception: 예외처리 구문이 작동되다가 오류가 발생하면 해당 오류 메시지를 출력
# as e: 해당 구문(Exception)을 축약하여 지정한 매개변수(e)를 사용
    print(e)
# 앞선 모든 구문이 끝나고 최종적으로 작동하는 구문
finally:
    print(data14)
```

- **재귀함수** (recursion) : 함수 내에서 자기 함수 스스로가 참조되어 작동하는 함수
    - 기저 조건(Base case) : 재귀함수 내에서 자기 자신이 호출되지 않도록 하는 조건
    - 기저 조건이 따로 없다면 무한재귀로 함수가 지속적으로 수행

```python
def func_recurs1(x):
    # 기저 조건 (x값이 반복되며 더해지다가, 10이 되는 순간 재귀가 정지
    if x == 10:
        print('End')
    else:
        print(x)
        func_recurs1(x+1)
```

```python
func_recurs1(1)
```

```python
# 람다 함수 : 함수를 한줄로 간결하게 만들 때 사용
# 함수의 결과를 변수로 선언하여 활용
def func20(a,b):
    return a + b
```

```python
data15 = func20(10,20)
data15
```

```python
# 위의 식보다 훨씬 간결해짐
func21 = lambda a,b : a+b
func21(30,40)
```

```python
func22 = lambda a,b : a.append(b)
list1 = [20, 30, 40]
func22(list1, 500)
```

# 4-4. 클래스 Class

- 특정 함수를 반복하여 사용할 때, 함수의 결과값을 계속 유지한 채로 계산할 수 없음
- 함수를 여러개의 변수로 선언하여 손쉽게 사용 -> 클래스
- 프로그래밍에서는 (개발) 클래스가 많이 사용이 됨
- 라이브러리 개념을 이해하기 위해 설명

```python
def func30(x):
    global result # global : 특정 변수가 함수내에서만 선언되지 않고, 전체 코드에서 사용 가능하게끔 바꿔주는 기능
    result += x
    return result
```

```python
result = 0
```

```python
func30(100)
```

```python
func30(300)
```

```python
func30(600)
```

```python
func30(0)
```

```python
class add_class():
    # 초기화 함수를 설정
    # 특정값의 합을 계산하는 변수를 0으로 초기화
    # 앞서 함수에서 사용한 glabal 역할 수행
    def __init__(self):
        self.result = 0
        #self : 클래스를 변수로 선언하여 사용할 때, 변수명에 자동으로 클래스 내의 함수를 전달함

    def func40(self, x):
        self.result += x
        return self.result
```

```python
funcA = add_class()
funcB = add_class()
```

```python
funcA.func40(100)
funcA.func40(200)
funcA.func40(300)
```

```python
funcB.func40(100)
```

```python
funcB.func40(100)
```

## 6-3. 정규표현식

- 정규표현식(Regular Expression)은 문자열을 처리할 때 특정 패턴을 정의하는데 사용되는 문자열
- 문자열 패턴을 사용하여, 문자열 검색, 대체, 추출 작업등을 수행
- 정규표현식은 특정 규칙을 가지고 있어, 다양한 문자와 연산자를 조합하여 복잡한 문자 패턴을 표현
- 크롤링 또는 자연어처리에서 데이터의 특정 패턴을 찾거나 추출할 때 사용

**메타 문자 (Meta Characters)**
- 문자가 가진 원래의 의미가 아닌 특별한 용도로 사용되는 문자
- 정규표현식에서는 대괄호 [] 안에 포함된 문자들 중 하나와 매치를 뜻함

```python
# . ^ $ +
import re
```

```python
# 대괄호 [ ] : 대괄호 내 포함된 문자들 중 하나와 매칭되는지 확인
pattern = r'[abc]'
# Raw string -> r 문자열 내에 매타 문자가 포함되어 있다
```

```python
text1= 'apple is 사과'
re.findall(pattern, text1)
```

```python
pattern = r'[a-z]' # 대괄호 내 두 문자 사이의 -기호를 사용하면 두 문자 사이의 범위를 지정
re.findall(pattern, text1)
# findall(): 문자열 전체에서 패턴과 일치하는 모든 부분을 찾아 리스트로 변환 
```

```python
pattern = r'[가-힣]'
text2 = "so you don't cry fo me 세월 지나도 난 변하지 않아~"
re.findall(pattern, text2)
```

```python
pattern = r'[가-힣]+' # +기호는 공백을 기준으로 문자의 패턴 확인 ->규칙이라 잘 안외워짐
re.findall(pattern, text2)
```

```python
# ^ : 해당 문자열을 제외한 나머지 패턴을 확인
pattern = r'[^가-힣]+'
re.findall(pattern, text2)
```

```python
pattern = r'[a-zA-z]+' # 모든 알파벳 표현 단어를 추출
re.findall(pattern, text2)
```

```python
pattern = r'\d' # \d : 모든 숫자
text3 = "1초라도 안보이면, 2이렇게 초조한데, 3삼초는 어떻게 기다려"
re.findall(pattern, text3)
```

```python
# 소괄호는 특정 패턴 앞이나 뒤의 값을 확인
pattern = r'(\d+)\원'
text4 = '2000원 짜리 아이스크림 4개 주세요. 500원 잔돈입니다.'
re.findall(pattern, text4)
```

```python
# re.sub() : 패턴과 일치하는 모든 부분을 다른 문자로 대치
text5 = "그댄 나의 챔피언 너와 나의 챔피언 우리 함께 외치면 승리하리라 한국"
pattern = r'챔피언'
replace_pattern = '챔피폰~'

re.sub(pattern, replace_pattern, text5)
```

```python
# re.split() : 특정 패턴을 기준으로 문자를 분할
pattern = r'\W+' # \W : 비단어 문자
re.split(pattern, text5)
```
